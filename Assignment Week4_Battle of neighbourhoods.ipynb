## Project description

Koku is a businessman who inherited a family business operating Indian restaurant based in south east asia. He is currently looking for opportunities to expand his business in North America. After narrow down his alternatives, he decided to choose either New York City or Toronto as his setting point for his business in this region.

Below are some points he used for consideration:
- Both Cities represents the financial capital of their countries.
- Both Cities are diverse (Consisting of multicultural etnic group)

For making a well informed decsion, he would like to compare the neighbourhood of both cities to have a closer look. This is to help his family to gain a better understanding the Geographical advantages of different districts (neighbourhoods) between this two cities.

## Discussion of the background

Data Source below were selected:
- Wikipedia demograhics data
- Foursquare location and common places data

First, a set of demograhics data of Toronto and New York City is extracted from Wikipedia, including the Borough, Neighbourhood and their map location join together in a single table. This can be done by either using pandas or Beautiful Soup Library

From the neighbourhood tabular results extracted New York and Toronto, a first visualization in the form of map could be seen, by utilizing foluim library. This provides a quick insight about how the neighbourhoods are located in the city.

Next, by using Foursquare API, the places in each neighbourhood are extracted. The ultimate goal for this part is to form a summary table of neighbourhood vs it's most common places to visit.

Finally, by using K-means clustering, 5 cluster shall be formed, to be able to examine each cluster and determine the discriminating venue categories that distinguish each cluster. Ultimately this can further narrow down and forecast in which neighbourhood the business could be more viable.
